```python
# importando as bibliotecas
import win32com.client as client
import pandas as pd
import datetime as dt
```

```python
# lendo o arquivo excel

tabela = pd.read_excel("Contas a Receber.xlsx")
display(tabela)

tabela.info()
```

```python
# verificar a data de hoje

hoje = dt.datetime.now()
print(hoje)
```

```python
# pegar apenas os clientes que estão devendo

devedores = tabela.loc[tabela['Status'] == 'Em aberto']

display(devedores)

devedores = devedores.loc[devedores['Data Prevista para pagamento'] < hoje]

display(devedores)
```

```python
# enviar o email
outlook = client.Dispatch('Outlook.Application')
```

```python
emissor = outlook.session.Accounts('email que envia')
```

```python
# mensagem
mensagem = outlook.CreateItem(0)
mensagem.Display()
mensagem.To = 'meuclientedevedor+194@gmail.com'
mensagem.Subject = 'Atraso no Pagamento'

mensagem.Body = '''
    
   

'''

mensagem._oleobj_.Invoke(*(64209,0,8,0,emissor))
mensagem.Save()
mensagem.Send()
```

```python
# pegando os valores dos devedores
dados = devedores[['Valor em aberto','Data Prevista para pagamento','E-mail','NF']].values.tolist()

print(dados)
```

```python
# repetição o email

for dado in dados:
    destinatario = dado[2]
    nf = dado[3]
    prazo = dado[1]
    prazo = prazo.strftime('%d/%m/%Y')
    valor= dado[0]
    
    mensagem = outlook.CreateItem(0)
    mensagem.Display()
    mensagem.To = destinatario
    mensagem.Subject = 'Atraso no Pagamento'

    mensagem.Body = f'''
    Prezado Cliente,

    Verificamos um atraso no pagamento referente a NF {nf} com vencimento em {prazo} e valor total de R${valor:.2f}.
    Gostaríamos de verificar se há algum problema que necessite de auxílio de nossa equipe. 

    Em caso de dúvidas, é só entrar em contato com nosso time atráves do e-mail tecobreiautomaticamentecompython@gmail.com

    
    Att,
    Lysia da Datasoft
    '''

    mensagem._oleobj_.Invoke(*(64209,0,8,0,emissor))
    mensagem.Save()
    mensagem.Send()
```
